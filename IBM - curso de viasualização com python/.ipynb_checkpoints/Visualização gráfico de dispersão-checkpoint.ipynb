# Curso de Visualização da IBM
## Grafico dispersão

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

## O arquivo para criação do dataframe e encontrada no link fornecido no formato xlsx
### O arquivo excel ate a linha 20 e ocupado por uma imagem, logo nos interessa buscar a partir da linha 20

```python
df_can = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
df_can.head()
```

## O comando drop retira as colunas que não serão necessarias

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head()
```

## Renomeia as colunas odName, AreaName, RegName
## Coloca a coluna renomeada country como indice do dataframe

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
#df_can.set_index('Country', inplace=True)
```

## Cria a variavel years com todas das colunas dos anos que vamos precisar para gerar o grafico de caixa

```python
years = list(map(int, range(1980, 2014)))
df_year = df_can[years].transpose()

df_year['total'] = df_year.sum(axis=1)
df_year.head()
```

```python
total = df_year[[1, 'total']]
#o ano deve ser colocado com o float para o uso da biblioteca sns
total.index = map(float, total.index)

total[1] = total.index
total.reset_index(inplace=True, drop=True)
```

```python
total.rename(columns={1:'ano'}, inplace=True)
```

```python
total.columns = ['ano', 'total']
```

```python
total.plot(kind='scatter', x='ano', y='total')

plt.title("dispersão de imigração por ano")
plt.ylabel("numero de imigrante")


plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))
sns.set(font_scale=1.5)
ax = sns.regplot(x='ano' , y ='total', data = total, color = 'orange', marker = '*', scatter_kws={'s': 200})
```
