## O arquivo para criação do dataframe e encontrada no link fornecido no formato xlsx
### O arquivo excel ate a linha 20 e ocupado por uma imagem, logo nos interessa buscar a partir da linha 20

```python
df_can = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
df_can.head()
```

## O comando drop retira as colunas que não serão necessarias

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head()
```

## Renomeia as colunas odName, AreaName, RegName 
## Coloca a coluna renomeada country como indice do dataframe

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.set_index('Country', inplace=True)
```

## adiciona a coluna 'total' ao dataframe

```python
df_can['Total'] = df_can.sum(axis=1)
df_can.sort_values(['Total'], ascending = False, axis = 0, inplace = True)

df_can_top5 = df_can.head(5)
```

## Cria a variavel years com todas das colunas dos anos que vamos precisar para gerar o grafico de area

```python
years = list(map(int, range(1980, 2014)))

#para usar o grafico de area a informação que queremos plotar deve estar na forma de coluna 
df_can_top5 = df_can_top5[years].transpose()

df_can_top5.plot(kind = 'area')

plt.title("top 5 paises com maior taxa de imigração para o canada")
plt.ylabel("numero de imigrantes")
plt.xlabel("Years")

plt.show()
```
